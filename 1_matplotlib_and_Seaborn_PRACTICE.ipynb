```python
import matplotlib.pyplot as plt
import numpy as np

g= np.array([35,5,25,35])
mylabels = ["Apples","Jackfriut","Dragonefriut","Kiwi"]

plt.pie(g,labels=mylabels)      #Label assigns the labels to chart values.
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

g= np.array([35,5,25,35])
mylabels = ["Apples","Jackfriut","Dragonefriut","Kiwi"]

plt.pie(g,labels=mylabels,startangle=110)                    #Graph is rotated bg given angle in anticlockwise direction.
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

g= np.array([35,5,25,35])
mylabels = ["Apples","Jackfriut","Dragonefriut","Kiwi"]
myexplode = [.5,0,0,.3]

plt.pie(g,labels=mylabels,explode =myexplode)          #Explodes the correspondig value to from the pie chart  wrt loction and value(example l=0 & v=.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

g= np.array([35,5,25,35])
mylabels = ["Apples","Jackfriut","Dragonefriut","Kiwi"]
myexplode = [.5,0,0,.3]

plt.pie(g,labels=mylabels,explode =myexplode,shadow=True)          #Explodes the correspondig value to from the pie chart  and gives shadows effects.
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

g= np.array([35,5,25,35])
mylabels = ["Apples","Jackfriut","Dragonefriut","Kiwi"]
mycolors = ["Blue","Red","Black","g"]

plt.pie(g,labels = mylabels,colors= mycolors)         #Assigne the colors on nchart manualy
plt.show()
```

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
 
 
# Creating dataset
np.random.seed(15)

data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]
 
fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])

# Creating plot
bp = ax.boxplot(data)
 
# show plot
plt.show()
```
