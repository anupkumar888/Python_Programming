<h2> Demo 1: Calculate Measures of Central Tendency using Python </h2>

In this demo, you will be shown how to calculate the measures of central tendency such as mean, median and mode.

```python
#Step1: Import the required library

import pandas as pd
import statistics 
```

```python
cars["mpg"].dtype
```

```python
#Step4: Calculate the mean
cars.mean()                 #mean of all columns using pandas
```

```python
cars["cyl"].mean()            #mean of a given column using pandas
```

```python
from statistics import mean

mean(cars["vs"])             #mean of a given column using pandas
```

```python
#Step5: Calculate the median
cars.median()             #median of all columns using pandas 
```

```python
#Step6: Calculate the mode
cars.mode()
```

The mode is calculated column-wise. In the above table, the mode values are displayed against the <b>'model'</b> column which is an unrequired descrepancy. So, we will remove that column to get a clear result.

```python
#cars1 = cars.drop('model', axis=1)
#cars1.head()
cars1 = cars.drop("model",axis=1)
cars1.head()
```

```python
#cars1['disp'].mode()  
cars1["disp"].mode()         #mode of a given column using pandas
```
