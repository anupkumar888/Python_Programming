## **Task 2**.Convert the created series into DataFrame and replace the null values with zeroes.

```python

import pandas as pd
   
# Initialise data to lists.
data =[{'Sam':60,'Jackson':74,'Ahree':85},

{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},

{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},

{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85, 'Gray':95},

{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87 }]

df = pd.DataFrame.from_records(data,index=['English', 'History','Geography','Mathematics','Science'])
print(df)
```

```python
# Replace the null values in dataset to zeroes

import numpy as np
df_new = df.replace(np.nan, 0)
print("The New DataFrame:\n",df_new)
```

## **Task 3**.Transpose the DataFrame and create a new column 'Average' and fill the values in it by

calculating the average of all subjects.

```python
import numpy as np
import pandas as pd

# Average of all the subjects
av_row = df_new.mean(axis=1)
print (av_row)
```

```python
# Using DataFrame.insert() to add a column

df_new.insert(7, "Average", (av_row), False)
print(df_new)
```

## **Problem Statement 3**: Write a Python program to create a series from 1 to 1000 and select
only numbers divisible by 7 and 17.

```python
# importing Pandas & numpy
import pandas as pd
import numpy as np
  
# numpy array
data=np.arange(1,1001)
  
# creating series
s = pd.Series(data)

nl=[]
for x in s:
    if (x%7==0) and (x%17==0):
        nl.append(str(x))
print (','.join(nl))
```
