### Business Scenario:

Employee retention is one of the biggest metrics that a company should have in mind when thinking of growth. Employee attrition is caused when the total strength of the company is greatly reduced as more employees leave the company than expected.

So, what is **Attrition**?
It is basically the turnover rate of employees in a particular organization

Reasons for **Attrition**:

- Employees looking for better opportunities
- A negative working environment
- Bad management
- Sickness of an employee
- Excessive working hours

#### **1. What is the Salary trend based on their Job Role?**

```python
f,ax=plt.subplots(figsize=(20,10))
sns.lineplot(x='JobRole', y='MonthlyIncome', data=df,ax=ax)
```

#### **2. Customize line plot by providing title**


```python
f,ax=plt.subplots(figsize=(20,10))

sns.lineplot(x='JobRole', y='MonthlyIncome', data=df,ax=ax).set(xlabel="JobRole",ylabel="Income",title="Line Plot-Job vs. Income")
```

#### **3. Visualize the trend based on Gender** 

```python
f,ax=plt.subplots(figsize=(20,10))

sns.lineplot(x='JobRole', y='MonthlyIncome',hue="Gender",err_style="bars",data=df,ax=ax).set(xlabel="JobRole",ylabel="Income",title="Line Plot-Job vs. Income")
```

### Inference:

1. Line plot usually shows the trend over time
2. We can also say that there is no Gender Discrimination towards Salary
3. However, Manager and Research Director earns the highest salaries. Sales Representatives, Research Scientist and Laboratory Technician earns the lowest salaries respectively

#### **10. What is Percentage of Research Scientist in IBM?**

```python
labels=df.JobRole.value_counts().index
colors=["olive","orange","hotpink","slateblue","y","lime"]
sizes=df.JobRole.value_counts().values
plt.figure(figsize=(8,8))
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%")
plt.title("Percentage of Research Scientists",color="saddlebrown",fontsize="20")
```

Parameters Used while plotting:

1. **Labels:** A sequence of strings providing the labels for each wedge
2. **Colors:** A sequence of matplotlibcolorargs through which the pie chart will cycle. If None, will use the colors in the currently active cycle
3. **Autopct:** Used to label the wedges with their numeric value. The label will be placed inside the wedge

Let us use explode to highlight only the Research Scientists from the chart

```python
labels=df.JobRole.value_counts().index
colors=["olive","orange","hotpink","slateblue","y","lime"]
explode=[0,0.3,0,0,0,0,0,0,0]
sizes=df.JobRole.value_counts().values
plt.figure(figsize=(8,8))
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%",explode=explode)
plt.title("Percentage of Research Scientists",color="saddlebrown",fontsize=20)
```

### Inference:

There are about 20% Research Scientists present in IBM. However, Sales Executives are more in Number
